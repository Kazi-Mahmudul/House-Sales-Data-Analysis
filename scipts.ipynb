#### Outlier identification and handling

An outlier is a data point that is significantly different from other observations. Detecting and handling outliers is essential in data analysis. 

#### Identification by Z-score

```python
# Z-score

import numpy as np

# Define function
def detect_outliers(data, threshold = 3 ):
    mean = np.mean(data)
    std = np.std(data)
    z_score = np.abs((data - mean)/ std)   # Z-score formula

    return z_score > threshold
        
```

The Z-score is widely used to method for identifying outliers. It measures how many standand daviations a data point is away from the mean. The formula for Z-score is (data - mean)/std. I defined detect_outlier as a function to detect the outliers. Set threshold=3 means 3rd standard deviation, it returns only that values which are bigger than the threshold.

#### Perform a T-Test 

A t-test is a statistical test used to campare the mean of two groups and determine if the differences between them are statistically significant.

#### Price difference based on bedrooms

```python
from  scipy.stats import ttest_ind

# Create two groups of bedrooms
group1 = df[df['bedrooms'] <= 3]['price']
group2 = df[df['bedrooms'] > 3]['price']

# Perform t-test
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

# Display results
print(f"t-stat: {t_stat: .4f}, p_value: {p_value: .4f}")

# Condition
if p_value < 0.05:
    print("Significant difference in price based on bedrooms")
else:
    print("No significant difference in price based on bedrooms")
```

This code is used to perform the t-test to determine the significant difference in price based on homes with <=3 and >3 bedrooms. ttest_ind() method is used to find the t-statistics and p_value of the two groups but the output shows the p_value is 0 which is extremely low.

```python
import numpy as np

# Cohen's d function
def cohen_d(group1, group2):
    mean_diff = np.mean(group1) - np.mean(group2)
    std_pool = np.sqrt((np.std(group1, ddof=1)**2 + np.std(group2, ddof=1)**2)/2)
    return mean_diff / std_pool

# Create two groups of bedrooms
group1 = df[df['bedrooms'] <= 3]['price']
group2 = df[df['bedrooms'] > 3]['price']

# d_value
d_value = cohen_d(group1, group2)
print(f"Cohen's d: {d_value: .3f}")

if abs(d_value) >= 0.5:
    print("Moderate effect")
elif abs(d_value) >= 0.3:
    print("Small effect")
else:
    print("High effect")
```

Because previously the p_value was 0, that's why I apply Cohen's d to measure effect size, to determine the magnitude of differences between the two groups. It is measured by dividing the mean differences by the pooled standard deviation. I applied some conditions to measure the effect size, and the output shows they have a moderate effect based on the conditions. 

#### Price difference based on total area
